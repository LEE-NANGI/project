```python
# Function to print the shape of the DataFrame
def data_shape(data):
    print("Data Shape:")
    print(f"Number of Rows: {data.shape[0]}")
    print(f"Number of Columns: {data.shape[1]}\n")

# Function to display information about the data
def data_info(data):
    print("Data Information:")
    print(data.info())

# Function to check for missing values
def data_missing(data):
    print("\nMissing Values:")
    print(data.isnull().sum())
# Function to identify and display duplicate rows
def identify_duplicates(data):
    duplicates = data[data.duplicated()]
    print("\nDuplicate Rows:")
    print(duplicates)
# Function to display descriptive statistics of numerical columns
def data_describe(data):
    print("\nDescriptive Statistics:")
    print(data.describe())
```

```python
# Function to explore the dataset
def explore_dataset(data):
    data_shape(data)
    data_info(data)
    data_missing(data)
    identify_duplicates(data)
    data_describe(data)
```

```python
explore_dataset(df)
```

```python
# Define a fuction to calculate the percentage of duplicates in each column of our dataset
def calculate_missing_percentage(data):
    # Calculate the total number of records
    total_records = len(data)

    # Check for missing values and calculate the percentage for each column
    missing_count = data.isnull().sum()
    missing_percentage = (missing_count / total_records) * 100

    # Create a DataFrame to store the missing values and percentages
    missing_data_df = pd.DataFrame({
        'Missing Values': missing_count,
        'Percentage (%)': missing_percentage
    })

    # Sort the DataFrame in ascending order based on the percentage of missing values
    missing_data_df = missing_data_df[missing_data_df['Missing Values'] > 0]
    missing_data_df.sort_values(by='Percentage (%)', ascending=True, inplace=True)

    return missing_data_df

# Assuming your dataset is named 'heart_data'
# Call the function to calculate missing percentages
missing_data_df = calculate_missing_percentage(df)

# Display the DataFrame with missing values and percentages
print(missing_data_df)
```

```python
# Define a function to drop missing values for columns with missing values less than 2%
def drop_missing_values(data, columns_to_drop):
    # Drop rows with missing values in the specified columns
    data.dropna(subset=columns_to_drop, inplace=True)

    return data

# Columns with missing values to be dropped
columns_to_drop = ["name", "host_name"]

# Adjust the column names in the 'columns_to_drop' list if needed
# For example, if there are leading or trailing whitespaces in the column names:
columns_to_drop = [col.strip() for col in columns_to_drop]

# Call the function to drop missing values in the specified columns
df = drop_missing_values(df, columns_to_drop)
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
# Handle glucose and eduaction columns
def handle_missing_values(data):
    # Replace missing values in the last_review with the last review date
    last_review = data['last_review'].max()
    df['last_review'].fillna(last_review, inplace=True)

    # Replace missing values in the reviews_per_month column with 0
    review_per_month = 0.0
    df['reviews_per_month'].fillna(review_per_month, inplace=True)

    return data

# Call the function to handle missing values
data = handle_missing_values(df)

# Verify if there are any missing values left after imputation
print(data.isnull().sum())
```

```python
relevant_columns = data2.drop(columns=['id', 'name', 'host_id', 'host_name', 'last_review', 'neighbourhood'])
relevant_columns.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_columns = relevant_columns.select_dtypes(include=['object'])
categorical_columns
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False)

encoded_categorical_columns = ohe.fit_transform(categorical_columns)
encoded_categorical_columns = pd.DataFrame(encoded_categorical_columns, columns=ohe.get_feature_names_out(categorical_columns.columns),  index=categorical_columns.index)
encoded_categorical_columns
```

```python
df_concat = pd.concat([relevant_columns.drop(columns=['neighbourhood_group', 'room_type']), encoded_categorical_columns], axis=1)
df_concat
```

```python
# df_concat.dropna(inplace=True)
# df_concat
```

```python
# Log-transform the target variable to address skewness
y = df_concat['price']
X = sm.add_constant(df_concat.drop(columns=['price']))


y_log = y.apply(lambda x: np.log1p(x))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.2, random_state=42)
# model = sm.OLS(y, X).fit()
model = sm.OLS(y_train, X_train).fit()
print(model.summary())
```

```python
# testing for normality using a qq plot. 
import scipy.stats as stats


fig, ax = plt.subplots()
sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True, ax=ax)

scatter = ax.lines[0]
line = ax.lines[1]
# scatter.set_markeredgecolor("blue")
# scatter.set_markerfacecolor("blue")
line.set_color("black")
fig.suptitle("");
```

```python
# Testing for normality using a histogram

fig, ax = plt.subplots()
sns.histplot(model.resid, bins=20, element="step", kde=True, ax=ax)
ax.set_xlabel("Model Residuals")
fig.suptitle("");
```

```python
# Testing for homoscedasticity using a residual Plot

residuals = model.resid
predicted = model.predict()

# residuals = actual_values - predicted_values

plt.scatter(predicted, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
